```python
import pandas as pd
dfp = pd.read_excel("c:/users/change/documents/monthly_production_05-2019.xlsx")
dfr = pd.read_excel("c:/users/change/documents/monthly_revenue_05-2019 (1).xlsx")
```

```python
dfr.describe()
```

```python
pro_vol = dfp["Volume"]
print("Low Range: " + str(pro_vol.mean() - pro_vol.std()))
print("High Range: " + str(pro_vol.mean() + pro_vol.std()))
print(pro_vol.min())
print(pro_vol.max())
```

```python
g = dfr.groupby(["Commodity"]) 
for c, d in g:
    print("-------\n" + str(c) + "\n-------")
    ind = d.index
    maxs = d["Revenue"].mean() + (d["Revenue"].std() * 3)
    mins = d["Revenue"].mean() - (d["Revenue"].std() * 3)
    
    for i in ind:
        value = dfr.loc[i,"Revenue"]
        if value >= maxs or value <= mins:
            print(i, value, sep=": ")
```

```python
g = dfp.groupby(["Commodity"]) 
for c, d in g:
    print("-------\n" + str(c) + "\n-------")
    ind = d.index
    maxs = d["Volume"].mean() + (d["Volume"].std() * 3)
    mins = d["Volume"].mean() - (d["Volume"].std() * 3)
    
    for i in ind:
        value = dfp.loc[i,"Volume"]
        if value >= maxs or value <= mins:
            print(i, value, sep=": ")
        
```

```python
import json
```

```python
class Car:
    __slots__ = ['wheels']
    
    def __init__(self, wheels):
        self.wheels = ['2','2']

hyundai = Car(4)

def encode_car(car):
    return car.wheels
```

```python
with open("somefile.json", "w") as file:
    json.dump(hyundai, file, indent=2, default=encode_car)
```
